```python
from selenium import webdriver
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import TimeoutException
from bs4 import BeautifulSoup
import time
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException, ElementNotInteractableException
from time import sleep

driver=webdriver.Chrome()
driver.maximize_window()

driver.get("https://www.instagram.com")
wait= WebDriverWait(driver,10)

```

# Problem 1 : Login to your Instagram Handle

```python

def LogIn(username, password):
    # Locating username textbox and sending username
    user_name = wait.until(EC.presence_of_element_located((By.NAME, "username")))
    user_name.send_keys(username)

    # Locating password box and sending password
    password_box = wait.until(EC.presence_of_element_located((By.NAME, "password")))
    password_box.send_keys(password)

    # Locating login button using correct locator
    button = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="loginForm"]/div[1]/div[3]/button')))
    button.click()

    # Handling potential pop-up (Ensure the locator is correct)
    try:
        pop = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="react-root"]/section/main/div/div/div/button')))
        pop.click()
    except:
        print("No pop-up appeared")
```

# Search Box


```python
def Search(string):
    try:
        print("Clicking search bar...")
        wait.until(EC.presence_of_element_located((By.XPATH, "//span[text()='Search']"))).click()

        print("Entering text...")
        search_input = wait.until(EC.presence_of_element_located((By.XPATH, "//input[@aria-label='Search input']")))
        search_input.send_keys(string)

        print("Fetching results...")
        data = wait.until(EC.presence_of_all_elements_located((
            By.XPATH, "//a[starts-with(@href, '/') and not(contains(@href, '?')) and not(contains(@href, 'logout'))]"
        )))
        print("success")

        print("Extracting usernames...")
        usernames = set()  # Use a set to avoid duplicates

        for i in data:
            href = i.get_attribute('href')
            if href and 'instagram.com' in href:
                username = href.rstrip('/').split('/')[-1]  # Extract username from the URL
                if username and not username.startswith('#'):
                    usernames.add(username)

        for name in usernames:
            print(name)

        sleep(5)
        print("Done.\n")

    except TimeoutException as e:
        print("Something Went Wrong!", str(e))
```

```python
Search('food')
```

```python
def Search_and_Open_Profile(string):
    try:
        # Locating Search bar then click on it
        wait.until(EC.presence_of_element_located((By.XPATH,"//span[text()='Search']"))).click()
        sleep(2)
        # Locating Search TextBox then enter search string
        wait.until(EC.presence_of_element_located((By.XPATH,"//input[@aria-label='Search input']"))).send_keys(string)
        sleep(5)
        # Find all Search handle and click on it
        wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'xuxw1ft'))).click()

    except TimeoutException:
        print("Something Went Wrong!")
```

# FOllOW AND UNFOLLOW

```python
def follow():
    try:
        # Locate the Follow button
        handle = wait.until(EC.presence_of_element_located((By.XPATH,"//header//button")))

        # Check if the button text is exactly "Follow"
        if handle.text.strip().lower() == "follow":
            sleep(3)
            handle.click()
            print("Followed user!")
        else:
            print("Already following or not a Follow button.")
            
    except TimeoutException:
        print("Something Went Wrong!")
```

```python
def unfollow():
    try:
        # Wait for the main follow/following button
        handle = wait.until(EC.presence_of_element_located((By.XPATH, "//header//button")))
        button_text = handle.text.strip().lower()
        print(f"Button text found: '{button_text}'")

        # If it's already unfollowed
        if "follow" in button_text and "following" not in button_text:
            print("Already unfollowed.")
        else:
            # Click the "Following" button to trigger the confirmation modal
            handle.click()
            sleep(1)

            # Confirm unfollow in modal (span text inside button)
            confirm = wait.until(EC.presence_of_element_located((
                By.XPATH, "//span[normalize-space()='Unfollow']"
            )))
            confirm.click()
            print("Unfollowed successfully.")

    except TimeoutException:
        print("Something went wrong while unfollowing.")
```

```python
Search_and_Open_Profile('So Delhi')
sleep(5)

# Following it
follow()
sleep(5)

# Unfollowing it
unfollow()
```

# Like And Unlike Post 
Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.
Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it


```python
def Liking():
    try:
        handle =wait.until(EC.presence_of_element_located((By.CLASS_NAME,"_aagw")))
        handle.click()
        #locating the text button
        Likebutton=wait.until(EC.presence_of_element_located((By.XPATH,"//section[1]/span[1]/div/div")))
        GetText = BeautifulSoup(Likebutton.get_attribute('innerHTML'),"html.parser").svg['aria-label']
        if GetText=='Like':
            # if not Liked then like it
            Likebutton.click()
            time.sleep(2)
        else:
            print('Already liked it')
            
        #locating Next post Button
        wait.until(EC.presence_of_element_located((By.XPATH,"//button[contains(@class, '_abl-')]"))).click()
        sleep(2)
        for i in range(2,30):
            Likebutton=wait.until(EC.presence_of_element_located((By.XPATH,"//article//section[1]//span[1]//div/div")))
            GetText = BeautifulSoup(Likebutton.get_attribute('innerHTML'),"html.parser").svg['aria-label']
            if GetText=='Like':
                # if not Liked then like it
                Likebutton.click()
                time.sleep(2)
            else:
               print('Already liked it')
            wait.until(EC.presence_of_element_located((By.XPATH,"/html/body/div[5]/div[1]/div/div[3]/div/div/div/div/div[1]/div/div/div[2]/button"))).click()
        #when 30 post completed go back to home page
        driver.back()

    except TimeoutException:
        print("Something Went Wrong!")
```

```python
def Unliking():
    try:
        handle =wait.until(EC.presence_of_element_located((By.CLASS_NAME,"_aagw")))
        handle.click()
        #locating the text button
        unlikebutton=wait.until(EC.presence_of_element_located((By.XPATH,"//section[1]/span[1]/div/div")))
        GetText = BeautifulSoup(unlikebutton.get_attribute('innerHTML'),"html.parser").svg['aria-label']
        if GetText=='Unlike':
            # if not Liked then like it
            unlikebutton.click()
            time.sleep(2)
        else:
            print('Already Unliked it')
            
        #locating Next post Button
        wait.until(EC.presence_of_element_located((By.XPATH,"//button[contains(@class, '_abl-')]"))).click()
        sleep(2)
        for i in range(2,30):
            unlikebutton=wait.until(EC.presence_of_element_located((By.XPATH,"//article//section[1]//span[1]//div/div")))
            GetText = BeautifulSoup(unlikebutton.get_attribute('innerHTML'),"html.parser").svg['aria-label']
            if GetText=='Unlike':
                # if not Liked then like it
                unlikebutton.click()
                time.sleep(2)
            else:
               print('Already Unliked it')
            wait.until(EC.presence_of_element_located((By.XPATH,"/html/body/div[5]/div[1]/div/div[3]/div/div/div/div/div[1]/div/div/div[2]/button"))).click()
        #when 30 post completed go back to home page
        driver.back()

    except TimeoutException:
        print("Something Went Wrong!")
```

```python
# Search and open profile of given Holder Id
Search_and_Open_Profile('dilsefoodie')
sleep(1)

# liking the post of dilsefoodie
Liking()

sleep(1)

Search_and_Open_Profile('dilsefoodie')
sleep(1)
# unliking the post of dilsefoodie
Unliking()
```

# Problem 6 : Extract list of followers

```python
def Extract_list_of_followers():
    try:
        print("[1] Clicking on the followers count...")
        followers_btn = wait.until(EC.element_to_be_clickable((By.XPATH, '//ul/li[2]//span')))
        followers_btn.click()
        print("Clicked followers count")

        print("[2] Waiting for followers scrollable frame to appear...")
        frame = wait.until(EC.presence_of_element_located((
            By.XPATH, '//div[contains(@class, "x1n2onr6") and contains(@class, "x1rife3k")]'
        )))
        print("Found followers modal scroll frame")

        print("[3] Scrolling the followers list...")
        for i in range(50):
            driver.execute_script('arguments[0].scrollTop = arguments[0].scrollHeight', frame)
            sleep(2)
        print("Scrolling done")

        print("[4] Extracting usernames...")
        handles = driver.find_elements(By.XPATH, '//div[@role="dialog"]//span[contains(text(), "")]')

        All_User = []
        for i in handles[:500]:
            username = i.text.split('\n')[0]
            if username:
                All_User.append(username)
        print(f"Extracted {len(All_User)} usernames")

        print("[5] Closing the modal...")
        driver.find_element(By.CLASS_NAME, '_abl-').click()
        print(" Modal closed")

        return All_User

    except TimeoutException as e:
        print("TimeoutException: Element not found.")
        print("Details:", str(e))
        return []

    except Exception as e:
        print("General Exception:", str(e))
        return []

```

```python
Search_and_Open_Profile('foodtalkindia')
sleep(3)

followers_of_foodtalkindia = Extract_list_of_followers()
for count, user in enumerate(followers_of_foodtalkindia, start=1):
    print(count, user)
```

# Print all the followers of “foodtalkindia” that you are following but those who don’t follow you.

```python
def Following():

    try:
    
        # Locating the Following button and click on it
        wait.until(EC.presence_of_element_located((By.XPATH,'//section/main//header//ul/li[3]//a/span'))).click()

        # Locating Following List Frame
        frame = wait.until(EC.presence_of_element_located((By.XPATH,'//div[contains(@class, "x1n2onr6") and contains(@class, "x1rife3k")]')))

        for i in range(2):
            driver.execute_script('arguments[0].scrollTop=arguments[0].scrollHeight',frame)

        handle = wait.until(EC.presence_of_all_elements_located((By.XPATH,'//div[@role="dialog"]//span[contains(text(), "")]')))

        All_User=[]
        # iterate over handle
        for i in handle:
            TEXT = i.text.split('\n')
            if TEXT[-1]=='Following':
                All_User.append(TEXT[0])

        sleep(1)
        # Locating Cross Button and click on 
        driver.find_element(By.CLASS_NAME, '_abl-').click()

        return All_User
    except TimeoutException:
        print("Something Went Wrong!")
```

```python
# Search for Given Insta Holder Id and open his profile
Search_and_Open_Profile('foodtalkindia')
sleep(3)

# Extract Followers of foodtalkindia using Extract_list_of_followers() function
followers_of_foodIN=Extract_list_of_followers()

# Convert list into set
followers_of_foodIN = set(followers_of_foodIN)
```

```python
def check_story():
    try: 

        
        # Locating the story
        story = wait.until(EC.presence_of_element_located((By.XPATH,"//img[contains(@alt, 'profile')]")))
        
        #check the Profile photo size to find out story is available or not
        Height = wait.until(EC.presence_of_element_located((By.XPATH,''))).get_attribute('height')
        
        if Height=='168':
            print("Viewing the story")
            story.click()
        else:
            print('Already seen the story')
    except:
        
        # Getting error it means No story is available
        print('No story is available to view')
```

```python
def check_story():
    try:
        # Wait for and locate the story/profile image element
        story = wait.until(EC.presence_of_element_located((By.XPATH, "//canvas[contains(@aria-label, 'profile picture')] | //img[contains(@alt, 'profile')]")))
        
        # Get the height of the story/profile image
        height = story.get_attribute('height')
        
        # Compare height to determine if story is available
        if height == '168':
            print("Viewing the story")
            story.click()
        else:
            print('Already seen the story')
    except Exception as e:
        print('No story is available to view')
        print(f"Error: {e}")
```
